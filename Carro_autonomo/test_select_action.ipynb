# Debug pipeline para select_action()
Use este notebook para montar um estado sintético, rodar `select_action()` e inspecionar as saídas.

```python
import torch

from ai import Dqn

torch.manual_seed(0)

input_size = 5
nb_action = 3
gamma = 0.9

dqn = Dqn(input_size=input_size, nb_action=nb_action, gamma=gamma)

state = torch.randn(1, input_size)
state
```

```python
actions = [dqn.select_action(state, debug = True) for i in range(5)]
actions
```
