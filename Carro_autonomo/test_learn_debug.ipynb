# Debug pipeline para learn()
Este notebook cria dados sinteticos e roda `learn(..., debug=True)` para inspecionar os valores intermediarios.

```python
import torch

from ai import Dqn

torch.manual_seed(0)

input_size = 5
nb_action = 3
gamma = 0.9
batch_size = 4

dqn = Dqn(input_size=input_size, nb_action=nb_action, gamma=gamma)

batch_state = torch.randn(batch_size, input_size)
batch_next_state = torch.randn(batch_size, input_size)
batch_action = torch.randint(0, nb_action, (batch_size,), dtype=torch.long)
batch_reward = torch.randn(batch_size)
```

```python
dqn.learn(
    batch_state=batch_state,
    batch_next_state=batch_next_state,
    batch_reward=batch_reward,
    batch_action=batch_action,
    debug=True,
)
```
