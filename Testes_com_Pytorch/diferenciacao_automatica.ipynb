Nesse notebook estarei testando a diferenciação automatica com pytorch:

```python
#Definição de uma função a ser diferenciada:
def foo(x, y):
    return x**2 + y**3
```

```python
#Criação dos tensores:
import torch

x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

x,y
```

```python
fc = foo(x, y)
fc
```

```python
fc.backward()
```

df/dx = 2*x; df/dy = 3*y^2

df/dx(2.0) = 4.0; df/dy(3.0) = 27

```python
print("df/dx(2.0): " + str(x.grad))
print("df/dy(3.0): " + str(y.grad))
```
