Neste notebook estarei implementando uma rede neural para prever um atributo (Income), dentro de um conjunto de atributos de um censo.

```python
import pandas as pd
base = pd.read_csv('census.csv')
base.head()
```

```python
base.income
```

Existem dois possíveis valores para income:

```python
base.income.unique()
```

```python
def converte_classe(rotulo):
    if rotulo == ' >50K':
         return 1
    return 0
```

```python
base.income = base.income.apply(converte_classe)
```

```python
base.head()
```

Definição dos inputs e output:

```python
X = base.drop('income', axis = 1)
X.head()
```

```python
Y = base.income
Y.head()
```

Separa-se a base de dados entre treino e teste:

```python
from sklearn.model_selection import train_test_split

X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size= 0.3)

X_treinamento.shape, X_teste.shape, Y_treinamento.shape, Y_teste.shape
```
